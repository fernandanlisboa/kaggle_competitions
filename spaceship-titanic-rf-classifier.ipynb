```python
# selecting columns
columns = df_train.columns.tolist()
# ['HomePlanet', 'Destination', 'Age', 'VIP', 'RoomService', 'Transported']
train = df_train[columns]
test = df_test[columns[:-1]]
```

### Working on categorical data

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
def grouping_categories(df_, col, labels, division=3):
    df = df_.copy(deep=True)
    min_value = df[col].min()
    max_value = df[col].max()
    step = np.round(max_value/division)
    intervals = np.arange( np.round(max_value), min_value, (-step), dtype=int)
    
    for i in range(len(labels)):
        df.loc[(df[col] < intervals[i]), f'{col}_Group'] = labels[i]
    
    return df
```

```python
def labeling_data(df, columns):
    df_ = df.copy()
    for col in columns:
        if df_[col].dtypes != 'int64' and df_[col].dtypes != 'int32':
            le = LabelEncoder()
            df_[col] = le.fit_transform(df_[col])
    return df_
```

```python
# grouping intervals
ages = ['Old', 'Adult', 'Young']
rooms = ['I', 'II', 'III', 'IV']
cols_todrop = ['Age', 'RoomService']

train = grouping_categories(train, cols_todrop[0], ages)
test = grouping_categories(test, cols_todrop[0], ages)
train = grouping_categories(train, cols_todrop[1], rooms, division=4)
test = grouping_categories(test, cols_todrop[1], rooms, division=4)

train.drop(columns=cols_todrop, inplace=True)
test.drop(columns=cols_todrop, inplace=True)
```

```python
# labeling categorical values
for c in cols_todrop:
    columns.remove(c)
    columns.append(f'{c}_Group')
labeled_train = labeling_data(train, columns)
labeled_test = labeling_data(test, test.columns)
```

### Separating data

```python
from sklearn.model_selection import train_test_split
```

```python
y = labeled_train['Transported']
X = labeled_train.drop(columns=['Transported'])
x_test = labeled_test.copy()

x_train, x_val, y_train, y_val = train_test_split(X, y, test_size=0.22, random_state=42)
print(f'x_train: {x_train.shape[0]}')
print(f'x_val: {x_val.shape[0]}')
print(f'x_test: {x_test.shape[0]}')
```

### Building Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_squared_error
```

```python
model = RandomForestClassifier(max_depth=15, n_estimators=500)
```

```python
# training
model.fit(x_train, y_train)
```

```python
model.score(x_val, y_val)
```

```python
y_pred = model.predict(x_val)

mean_squared_error(y_val, y_pred)
```

### Making prediction

```python
preds = model.predict(x_test)
output = pd.DataFrame({'PassengerId': x_test.index.tolist(), 'Transported': preds.tolist()})
output['Transported'] = output['Transported'].apply(lambda x: bool(1) if x >= 0.5 else bool(0))
output.to_csv('submission.csv', index=False)
print(output)
```
